Primero que nada tenemos que traer la informacion de la url proporcionada


```python
# Paso 1: Descargar y descomprimir el ZIP directamente desde la UCI
import urllib.request
import zipfile
import os

# Crear carpeta temporal
os.makedirs("sgc_data", exist_ok=True)

# URL del archivo ZIP
zip_url = "https://archive.ics.uci.edu/static/public/522/south+german+credit.zip"
zip_path = "sgc_data/south_german_credit.zip"

# Descargar el archivo
urllib.request.urlretrieve(zip_url, zip_path)

# Descomprimir
with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall("sgc_data")

# Verificar archivos extraídos
print("Archivos extraídos:", os.listdir("sgc_data"))
```

Posteriormente Cargamos el archivo y renombramos columnas

```python
import pandas as pd

# Ruta del archivo .asc extraído
file_path = "sgc_data/SouthGermanCredit.asc"

# Cargar el archivo
df = pd.read_csv(file_path, sep='\s+')

# Diccionario de renombre de columnas del alemán al inglés
column_translation = {
    'laufkont': 'status_of_existing_checking_account',
    'laufzeit': 'duration_in_month',
    'moral': 'credit_history',
    'verw': 'purpose',
    'hoehe': 'credit_amount',
    'sparkont': 'savings_account_bonds',
    'beszeit': 'present_employment_since',
    'rate': 'installment_rate_in_percentage_of_disposable_income',
    'famges': 'personal_status_and_sex',
    'buerge': 'other_debtors_or_guarantors',
    'wohnzeit': 'present_residence_since',
    'verm': 'property',
    'alter': 'age_in_years',
    'weitkred': 'other_installment_plans',
    'wohn': 'housing',
    'bishkred': 'number_of_existing_credits_at_this_bank',
    'beruf': 'job',
    'pers': 'number_of_people_being_liable_to_provide_maintenance_for',
    'telef': 'telephone',
    'gastarb': 'foreign_worker',
    'kredit': 'creditworthy'
}

# Renombrar columnas
df.rename(columns=column_translation, inplace=True)

# Mostrar primeras filas
df.head()
```

Realiza una partición de los datos en el conjunto de entrenamiento del 85% y el de prueba de
15%. Los modelos se estarán entrenando con el método de validación cruzada, así que no es

```python
from sklearn.model_selection import train_test_split

# Definir variables X e y
X = df.drop("creditworthy", axis=1)
y = df["creditworthy"]

# Partición 85% entrenamiento, 15% prueba
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.15, random_state=42, stratify=y
)

# Mostrar tamaño de cada conjunto
print("Tamaño de entrenamiento X:", X_train.shape)
print("Tamaño de prueba X:", X_test.shape)
print("Tamaño de entrenamiento y:", y_train.shape)
print("Tamaño de prueba y:", y_test.shape)
```

Categoricas->purpose, personal_status_and_sex, other_debtors_or_guarantors, property, other_installment_plans, housing, job

Binary:telephone, foreign_worker, creditworthy

Estrategias de imputacion.

Quantitativos->por la mediana  y la escogi porque suelen ser robustas por ser variables numericas.

Ordinales->Por moda y la escogi porque no tiene valores continuos.

Categoricas->moda,por ser mas valores nominales

Binarias->most_frequent por valores que son nominales

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
```

```python
# Listas de variables por tipo
numeric_features = ['duration_in_month', 'credit_amount', 'age_in_years']
ordinal_features = [
    'status_of_existing_checking_account',
    'credit_history',
    'savings_account_bonds',
    'present_employment_since',
    'installment_rate_in_percentage_of_disposable_income',
    'present_residence_since',
    'number_of_existing_credits_at_this_bank',
    'number_of_people_being_liable_to_provide_maintenance_for'
]
categorical_features = [
    'purpose',
    'personal_status_and_sex',
    'other_debtors_or_guarantors',
    'property',
    'other_installment_plans',
    'housing',
    'job'
]
binary_features = ['telephone', 'foreign_worker']

# Definir los imputers para cada tipo
numeric_imputer = SimpleImputer(strategy='median')
ordinal_imputer = SimpleImputer(strategy='most_frequent')
categorical_imputer = SimpleImputer(strategy='most_frequent')
binary_imputer = SimpleImputer(strategy='most_frequent')

# Construir el ColumnTransformer
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_imputer, numeric_features),
        ('ord', ordinal_imputer, ordinal_features),
        ('cat', categorical_imputer, categorical_features),
        ('bin', binary_imputer, binary_features)
    ]
)

# Pipeline completo (solo imputación en este paso)
pipeline = Pipeline(steps=[('preprocessor', preprocessor)])

# Aplicar el pipeline sobre el conjunto de entrenamiento
X_train_imputed = pipeline.fit_transform(X_train)

# También puedes transformar el conjunto de prueba después
X_test_imputed = pipeline.transform(X_test)

# Mostrar tipo y tamaño resultado
print(type(X_train_imputed))
print(X_train_imputed.shape)
```
